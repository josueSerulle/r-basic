Pista: investiga el parámetro fill_between de matplotlib

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.figure(figsize=(8, 5))

plt.plot(X, Y + 1, color="blue", alpha=1.0)
plt.plot(X, Y - 1, color="blue", alpha=1.0)

plt.fill_between(X, 1, Y + 1, color="blue", alpha=0.3)

plt.fill_between(X, -1, Y - 1, color="red", alpha=0.3)

plt.show()
```

## Pregunta 2

A partir del siguiente código, intenta obtener el gráfico del dibujo.

n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
plt.scatter(X, Y)
Pista: el color viene dado por el ángulo que forman x e y con el eje horizontal.


```python
n = 1024

X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

T = np.arctan2(Y, X)

plt.figure(figsize=(8, 6))
plt.scatter(X, Y, c=T, cmap='hsv', alpha=0.5)

plt.show()
```

n = 12

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, facecolor = "#9999ff", edgecolor = "white")

, +1.25)
Pista: cuidado con la alineación del texto en el centro de las barras..

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor="#9999ff", edgecolor="white")
plt.bar(X, -Y2, facecolor="#ff9999", edgecolor="white")

for x, y in zip(X, Y1):
    plt.text(x, y + 0.03, "%.2f" % y, ha="center", va="bottom")

for x, y in zip(X, Y2):
    plt.text(x, -y - 0.05, "%.2f" % y, ha="center", va="top")

plt.ylim(-1.25, +1.25)

plt.show()
```

## Pregunta 4
A partir del siguiente código, intenta obtener el gráfico del dibujo

Z = np.random.uniform(0, 1, 20)
plt.pie(Z)
Pista: tendrás que modificar la Z.

```python
Z = np.random.uniform(0, 1, 20)

explode = [0.1 if i == np.argmax(Z) else 0 for i in range(len(Z))]

colors = plt.cm.Greys(np.linspace(0, 1, len(Z)))

plt.figure(figsize=(8, 6))
plt.pie(Z, colors=colors, explode=explode, wedgeprops={"edgecolor": "white"})

plt.show()
```
